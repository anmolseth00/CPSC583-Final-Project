```python
import numpy as np
import pandas as pd
import pickle
import glob
import os
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
import json

# Set paths
BASE_PATH = Path('raw_data')
OPENVACCINE_PATH = BASE_PATH / 'openvaccine_12x_dataset'
STANFORD_PATH = BASE_PATH / 'stanford-covid-vaccine'

def load_train_data():
    """Load and return the training data from Stanford COVID dataset"""
    try:
        # Read the file line by line
        with open(STANFORD_PATH / 'train.json', 'r') as f:
            data = []
            for line in f:
                try:
                    data.append(json.loads(line.strip()))
                except json.JSONDecodeError as e:
                    print(f"Error parsing line: {e}")
                    continue
        
        train_df = pd.DataFrame(data)
        print(f"Training data shape: {train_df.shape}")
        return train_df
    except Exception as e:
        print(f"Error loading training data: {e}")
        print("\nLet's inspect the file structure:")
        with open(STANFORD_PATH / 'train.json', 'r') as f:
            first_lines = [next(f) for _ in range(5)]
        print("First few lines of the file:")
        print("\n".join(first_lines))
        return None

def inspect_file_structure():
    """Inspect the structure of our data directories"""
    print("OpenVaccine Dataset Structure:")
    print("-" * 50)
    file_counts = {
        'bpp': len(list(OPENVACCINE_PATH.glob('*_bpp.npy'))),
        'loop': len(list(OPENVACCINE_PATH.glob('*_loop.p'))),
        'struc': len(list(OPENVACCINE_PATH.glob('*_struc.p')))
    }
    print("File counts:", file_counts)
    
    print("\nStanford COVID Dataset Structure:")
    print("-" * 50)
    for file in STANFORD_PATH.glob('*'):
        print(f"File: {file.name}")
        if file.suffix == '.json':
            try:
                with open(file, 'r') as f:
                    first_line = f.readline().strip()
                print(f"First line preview: {first_line[:100]}...")
            except Exception as e:
                print(f"Error reading file: {e}")

def load_sample_rna_data():
    """Load and analyze a single RNA sample"""
    # Get first bpp file
    first_bpp = next(OPENVACCINE_PATH.glob('*_bpp.npy'))
    id_prefix = first_bpp.stem.replace('_bpp', '')
    
    try:
        bpp_data = np.load(first_bpp)
        loop_data = pickle.load(open(OPENVACCINE_PATH / f"{id_prefix}_loop.p", 'rb'))
        struc_data = pickle.load(open(OPENVACCINE_PATH / f"{id_prefix}_struc.p", 'rb'))
        
        print(f"\nSample RNA ID: {id_prefix}")
        print(f"BPP shape: {bpp_data.shape}")
        print(f"Loop data type: {type(loop_data)}")
        print(f"Structure data type: {type(struc_data)}")
        
        return {'bpp': bpp_data, 'loop': loop_data, 'struc': struc_data}
    except Exception as e:
        print(f"Error loading sample RNA data: {e}")
        return None

# Main execution
print("Inspecting data structure...")
inspect_file_structure()

print("\nAttempting to load sample RNA data...")
sample_data = load_sample_rna_data()

print("\nAttempting to load training data...")
train_df = load_train_data()
```

```python
import numpy as np
import pandas as pd
import pickle
import json
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

def load_sequence_data(stanford_path):
    """Load sequence data from the Stanford dataset"""
    sequences = []
    with open(stanford_path / 'train.json', 'r') as f:
        for line in f:
            sequences.append(json.loads(line))
    return pd.DataFrame(sequences)

def analyze_rna(seq_id, sequence_df, structure_path):
    """Analyze RNA sequence and structure together"""
    # Format ID for structure files
    structure_id = f"id_{seq_id}"
    
    # Load structure files
    bpp_data = np.load(f"{structure_path}/{structure_id}_bpp.npy")
    with open(f"{structure_path}/{structure_id}_loop.p", 'rb') as f:
        loop_data = pickle.load(f)
    with open(f"{structure_path}/{structure_id}_struc.p", 'rb') as f:
        struc_data = pickle.load(f)
    
    # Get sequence information
    seq_info = sequence_df[sequence_df['id'] == structure_id].iloc[0]
    
    print(f"RNA Analysis for {structure_id}")
    print("-" * 50)
    print(f"Sequence length: {len(seq_info['sequence'])}")
    print(f"Base pair probability matrix shape: {bpp_data.shape}")
    print("\nSequence:")
    print(seq_info['sequence'])
    print("\nStructure (dot-bracket):")
    print(seq_info['structure'])
    print("\nPredicted loop type:")
    print(seq_info['predicted_loop_type'])
    
    # Create visualization
    fig = plt.figure(figsize=(15, 10))
    
    # Plot BPP matrix
    plt.subplot(2, 1, 1)
    sns.heatmap(bpp_data[0], cmap='YlOrRd')  # Using first prediction
    plt.title('Base Pair Probability Matrix')
    plt.xlabel('Sequence Position')
    plt.ylabel('Sequence Position')
    
    # Plot reactivity data
    plt.subplot(2, 1, 2)
    plt.plot(seq_info['reactivity'], label='Reactivity')
    plt.title('RNA Reactivity Profile')
    plt.xlabel('Sequence Position')
    plt.ylabel('Reactivity')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    return {
        'sequence_info': seq_info,
        'bpp': bpp_data,
        'loop': loop_data,
        'structure': struc_data
    }

# Set paths
BASE_PATH = Path('raw_data')
STANFORD_PATH = BASE_PATH / 'stanford-covid-vaccine'
STRUCTURE_PATH = BASE_PATH / 'openvaccine_12x_dataset'

# Load sequence data
print("Loading sequence data...")
seq_data = load_sequence_data(STANFORD_PATH)

# Analyze specific RNA
rna_id = "0a2bbe37e"
rna_data = analyze_rna(rna_id, seq_data, STRUCTURE_PATH)

print("\nStructure Predictions:")
print("Number of alternative structures:", len(rna_data['loop']))
for i in range(min(3, len(rna_data['loop']))):
    print(f"\nPrediction {i+1}:")
    print(rna_data['loop'][i])
```

```python
import numpy as np
import pandas as pd
import pickle
import json
from pathlib import Path
import glob
import re

def load_stanford_data(stanford_path):
    """
    Load and organize Stanford COVID vaccine dataset
    Returns a dataframe with all sequence information
    """
    print("Loading Stanford dataset...")
    sequences = []
    
    # Load training data
    with open(stanford_path / 'train.json', 'r') as f:
        for line in f:
            sequences.append(json.loads(line))
    
    # Load test data if available
    try:
        with open(stanford_path / 'test.json', 'r') as f:
            for line in f:
                entry = json.loads(line)
                entry['dataset'] = 'test'  # Mark test entries
                sequences.append(entry)
    except Exception as e:
        print(f"Note: Test data not loaded: {e}")
    
    # Convert to dataframe
    df = pd.DataFrame(sequences)
    
    # Add derived columns
    df['seq_length'] = df['sequence'].str.len()
    df['gc_content'] = df['sequence'].apply(lambda x: (x.count('G') + x.count('C')) / len(x))
    
    print(f"Stanford dataset loaded: {len(df)} sequences")
    return df

def load_openvaccine_data(openvaccine_path):
    """
    Load and organize OpenVaccine 12x dataset
    Returns a dataframe with all structure predictions
    """
    print("\nLoading OpenVaccine dataset...")
    data = []
    
    # Get all unique IDs
    bpp_files = glob.glob(str(openvaccine_path / '*_bpp.npy'))
    unique_ids = [re.search(r'id_[a-f0-9]+', f).group() for f in bpp_files]
    
    for id_prefix in unique_ids:
        try:
            # Load BPP data
            bpp = np.load(f"{openvaccine_path}/{id_prefix}_bpp.npy")
            
            # Load loop predictions
            with open(f"{openvaccine_path}/{id_prefix}_loop.p", 'rb') as f:
                loop = pickle.load(f)
                
            # Load structure predictions
            with open(f"{openvaccine_path}/{id_prefix}_struc.p", 'rb') as f:
                struc = pickle.load(f)
            
            # Create entry for this RNA
            entry = {
                'id': id_prefix,
                'num_predictions': len(loop),
                'sequence_length': bpp.shape[1],
                'bpp_shape': bpp.shape,
            }
            
            # Add individual predictions
            for i in range(len(loop)):
                entry[f'loop_pred_{i+1}'] = loop[i]
                entry[f'structure_pred_{i+1}'] = struc[i]
                
            # Add BPP statistics for each prediction
            for i in range(bpp.shape[0]):
                entry[f'bpp_mean_{i+1}'] = bpp[i].mean()
                entry[f'bpp_max_{i+1}'] = bpp[i].max()
                entry[f'bpp_nonzero_{i+1}'] = (bpp[i] > 0).sum()
            
            data.append(entry)
            
        except Exception as e:
            print(f"Error processing {id_prefix}: {e}")
    
    df = pd.DataFrame(data)
    print(f"OpenVaccine dataset loaded: {len(df)} structures")
    return df

def analyze_datasets(stanford_df, openvaccine_df):
    """
    Print analysis of both datasets
    """
    print("\nDataset Analysis:")
    print("-" * 50)
    
    # Stanford dataset analysis
    print("\nStanford Dataset:")
    print(f"Total sequences: {len(stanford_df)}")
    print("\nColumns available:")
    for col in stanford_df.columns:
        print(f"- {col}")
    
    print("\nSequence length statistics:")
    print(stanford_df['seq_length'].describe())
    
    # OpenVaccine dataset analysis
    print("\nOpenVaccine Dataset:")
    print(f"Total structures: {len(openvaccine_df)}")
    print("\nColumns available:")
    for col in openvaccine_df.columns:
        print(f"- {col}")
    
    # Check overlap between datasets
    stanford_ids = set(stanford_df['id'].str.replace('id_', ''))
    openvaccine_ids = set(openvaccine_df['id'].str.replace('id_', ''))
    common_ids = stanford_ids.intersection(openvaccine_ids)
    
    print(f"\nSequences with both structure and experimental data: {len(common_ids)}")

# Set paths
BASE_PATH = Path('raw_data')
STANFORD_PATH = BASE_PATH / 'stanford-covid-vaccine'
OPENVACCINE_PATH = BASE_PATH / 'openvaccine_12x_dataset'

# Load both datasets
stanford_df = load_stanford_data(STANFORD_PATH)
openvaccine_df = load_openvaccine_data(OPENVACCINE_PATH)

# Analyze the datasets
analyze_datasets(stanford_df, openvaccine_df)

# Display sample entries
print("\nSample Stanford entry:")
print(stanford_df.iloc[0])

print("\nSample OpenVaccine entry:")
print(openvaccine_df.iloc[0])
```

```python
import numpy as np
import pandas as pd
import pickle
import json
from pathlib import Path
import glob
import re

# Load functions remain the same as previous code...
# [Previous load_stanford_data and load_openvaccine_data functions here]

def merge_and_report_mismatches(stanford_df, openvaccine_df):
    """
    Merge the two dataframes and report unmatched sequences
    """
    # Standardize IDs
    stanford_df['clean_id'] = stanford_df['id'].str.replace('id_', '')
    openvaccine_df['clean_id'] = openvaccine_df['id'].str.replace('id_', '')
    
    # Find unmatched IDs
    stanford_ids = set(stanford_df['clean_id'])
    openvaccine_ids = set(openvaccine_df['clean_id'])
    
    stanford_only = stanford_ids - openvaccine_ids
    openvaccine_only = openvaccine_ids - stanford_ids
    common_ids = stanford_ids.intersection(openvaccine_ids)
    
    print(f"\nMerging Statistics:")
    print("-" * 50)
    print(f"Total Stanford sequences: {len(stanford_df)}")
    print(f"Total OpenVaccine structures: {len(openvaccine_df)}")
    print(f"Sequences in both datasets: {len(common_ids)}")
    print(f"\nUnmatched sequences:")
    print(f"Stanford only: {len(stanford_only)} sequences")
    print(f"OpenVaccine only: {len(openvaccine_only)} sequences")
    
    # Print unmatched IDs
    if len(stanford_only) > 0:
        print("\nStanford sequences without structure predictions:")
        print(sorted(list(stanford_only))[:10], "..." if len(stanford_only) > 10 else "")
        
    if len(openvaccine_only) > 0:
        print("\nOpenVaccine structures without experimental data:")
        print(sorted(list(openvaccine_only))[:10], "..." if len(openvaccine_only) > 10 else "")
    
    # Merge dataframes
    merged_df = pd.merge(stanford_df, openvaccine_df, on='clean_id', suffixes=('_stanford', '_openvaccine'))
    
    # Clean up
    merged_df = merged_df.drop(['id_stanford', 'id_openvaccine'], axis=1)
    
    print(f"\nFinal merged dataset: {len(merged_df)} sequences")
    return merged_df

# Set paths
BASE_PATH = Path('raw_data')
STANFORD_PATH = BASE_PATH / 'stanford-covid-vaccine'
OPENVACCINE_PATH = BASE_PATH / 'openvaccine_12x_dataset'

# Load both datasets
stanford_df = load_stanford_data(STANFORD_PATH)
openvaccine_df = load_openvaccine_data(OPENVACCINE_PATH)

# Merge and report mismatches
merged_df = merge_and_report_mismatches(stanford_df, openvaccine_df)

# Save merged dataset if needed
# merged_df.to_csv('merged_dataset.csv', index=False)
```

Now let's see if there is any overlap and if there's any 1-1 correspondence

```python
import pandas as pd

def find_matching_structure_column(merged_df):
    # We'll store results in a dictionary: {column_name: count_of_matches}
    match_counts = {f'structure_pred_{i}': 0 for i in range(1, 13)}
    
    # Also store the matched column per row
    matched_columns = []
    
    for idx, row in merged_df.iterrows():
        original_structure = row['structure']
        
        matched_col = None
        # Check each of the 12 predictions
        for i in range(1, 13):
            pred_col = f'structure_pred_{i}'
            predicted_structure = row[pred_col]
            
            if original_structure == predicted_structure:
                match_counts[pred_col] += 1
                matched_col = pred_col
                break
        
        # If no match found, matched_col remains None, but it should always find one
        matched_columns.append(matched_col)
    
    # Convert matched_columns to a series if needed
    merged_df['matching_structure_pred'] = matched_columns
    
    # Print summary
    print("Match counts per predicted structure column:")
    for col, count in match_counts.items():
        print(f"{col}: {count} matches")
    
    return merged_df

# Example usage (assuming you already have merged_df):
merged_df = find_matching_structure_column(merged_df)

# Now merged_df has a 'matching_structure_pred' column indicating which structure_pred matched
# Also printed are the counts of how many sequences matched each prediction column.
```

The expectation was that one particular predicted structure (say, from Vienna at 37C) would correspond exactly to the Kaggle-provided structure for all sequences. However, the distribution of matches suggests that the original Kaggle structure data might not have been generated from a single, uniform condition or tool for all sequences. Or perhaps a different tool was used?
